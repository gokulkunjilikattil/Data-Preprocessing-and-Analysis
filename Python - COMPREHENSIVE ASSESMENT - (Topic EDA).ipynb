# Identify the predominant age group among employees.

```python
bins = [20, 25, 30, 35, 40, 45, 50]
labels = ['20-24', '25-29', '30-34', '35-39', '40-44', '45-49']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
```

```python
age_group_distribution = data['Age Group'].value_counts().sort_index()
```

```python
age_group_distribution_df = pd.DataFrame({'Age Group': age_group_distribution.index,
                                          'Number of Employees': age_group_distribution.values})
```

```python
print(age_group_distribution_df)
```

```python
predominant_age_group = age_group_distribution.idxmax()
predominant_count = age_group_distribution.max()
```

```python
print(f"The predominant age group is {predominant_age_group} with {predominant_count} employees.")
```

```python
plt.figure(figsize=(12, 8))
age_group_distribution.plot(kind='bar')
plt.title('Distribution of Employees Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

# Discover which team and position have the highest salary expenditure. 

```python
team_salary_expenditure = data.groupby('Team')['Salary'].sum().sort_values(ascending=False)
```

```python
position_salary_expenditure = data.groupby('Position')['Salary'].sum().sort_values(ascending=False)
```

```python
highest_salary_team = team_salary_expenditure.idxmax()
highest_salary_team_amount = team_salary_expenditure.max()
```

```python
highest_salary_position = position_salary_expenditure.idxmax()
highest_salary_position_amount = position_salary_expenditure.max()
```

```python
print(f"The team with the highest salary expenditure is {highest_salary_team} with a total of ${highest_salary_team_amount:.2f}.")
print(f"The position with the highest salary expenditure is {highest_salary_position} with a total of ${highest_salary_position_amount:.2f}.")
```

```python
plt.figure(figsize=(12, 8))
team_salary_expenditure.plot(kind='bar')
plt.title('Total Salary Expenditure per Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
position_salary_expenditure.plot(kind='bar')
plt.title('Total Salary Expenditure per Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')
plt.show()
```

# Investigate if there's any correlation between age and salary, and represent it visually.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
correlation_coefficient = data['Age'].corr(data['Salary'])
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=data)
plt.title(f'Correlation Between Age and Salary: {correlation_coefficient:.2f}')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
print(f"The correlation coefficient between age and salary is {correlation_coefficient:.2f}.")
```
